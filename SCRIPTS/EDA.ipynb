```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

#creating data frame
df = pd.read_csv('../DATA/metadata.csv')
# isolating skin cancer counts
diagnostic_counts = df['diagnostic'].value_counts()
```

```python
#print the counts of each type
print(diagnostic_counts)
```

## Overall Distribution

```python
#gender graph
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df, palette=['lightblue', 'lightcoral'])
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender")
plt.show()
```

```python
# skin leison graph
plt.figure(figsize=(10, 6))
sns.countplot(x='region', data=df, palette='viridis')
plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Distribution of Regions")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Set the style and palette
sns.set_theme(style="whitegrid")
palette = sns.color_palette("pastel")

# Create a boxplot
sns.boxplot(data=df, x='age', palette=palette, showmeans=True, meanline=True, meanprops={"color": "red", "ls": "--", "lw": 2})

plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

## Six-Class Distribution

```python
plt.figure(figsize=(8, 6))

# Count the occurrences of each Category within each Type
counts = df.groupby(['diagnostic', 'gender']).size().reset_index(name='Count')

# Create barplot
sns.barplot(data=counts, x='diagnostic', y='Count', hue='gender', palette=['lightblue', 'lightcoral'])

plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender by Diagnostic")
plt.show()
```

```python
# Count the occurrences of each Category within each Type
counts = df.groupby(['diagnostic', 'region']).size().reset_index(name='Count')

# Create barplot
sns.barplot(data=counts, x='diagnostic', y='Count', hue='region', palette='viridis')

# Customize the plot
plt.title('Distribution of Regions by Diagnostic')
plt.xlabel('Diagnostic')
plt.ylabel('Count')
plt.legend(title='Region')
plt.show()

```

```python
# Set the style and palette
sns.set_theme(style="whitegrid")
palette = sns.color_palette("pastel")

# Create a boxplot
sns.boxplot(data=df, x='age', y="diagnostic", palette=palette, showmeans=True, meanline=True, meanprops={"color": "red", "ls": "--", "lw": 2})

plt.title('Age Distribution by Diagnostic')
plt.xlabel('Age')
plt.ylabel('Diagnostic')
plt.show()
```

## Two-Class Distribution

```python
mapping = {'BCC':'MALIGNANT', 'MEL':'MALIGNANT', 'SCC':'MALIGNANT' , 'ACK':'BENIGN', 'NEV':'BENIGN', 'SEK':'BENIGN'}
df['class'] = df['diagnostic'].map(mapping)
```

```python
# Count the occurrences of each Category within each Type
counts = df.groupby(['class', 'gender']).size().reset_index(name='Count')

# Create barplot
sns.barplot(data=counts, x='class', y='Count', hue='gender', palette=['lightblue', 'lightcoral'])

plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender by Tumor Type")
plt.show()
```

```python
# Count the occurrences of each Category within each Type
counts = df.groupby(['class', 'region']).size().reset_index(name='Count')

# Create barplot
sns.barplot(data=counts, x='class', y='Count', hue='region', palette='viridis')

# Customize the plot
plt.title('Distribution of Regions by Tumor Type')
plt.xlabel('Tumor Type')
plt.ylabel('Count')
plt.legend(title='Region')
plt.show()
```

```python
palette = sns.color_palette("pastel")

# Create a boxplot
sns.boxplot(data=df, x='age', y="class", palette=palette, showmeans=True, meanline=True, meanprops={"color": "red", "ls": "--", "lw": 2})

plt.title('Age Distribution by Tumor Type')
plt.xlabel('Age')
plt.ylabel('Tumor Type')
plt.show()
```
